```python
#Q2. Max, min , mean of asked amount,asked equity, asked valuation?
df[['pitcher_ask_amount', 'ask_equity', 'ask_valuation']].agg(['max','mean','min']).T
```

```python
#Q3. Max & Min asked- equity,asked-valuation and asked amount episode-wise?
df.groupby('episode_number')[['pitcher_ask_amount', 'ask_equity', 'ask_valuation']].agg(['max','min'])
```

```python
df['total_sharks_invested']>=2
```

```python
#Q4. Brand names in which 2 ,3 or 4 sharks are invested?
df[(df['total_sharks_invested']>=2 )& (df['total_sharks_invested']<=4)][['brand_name']]
```

```python
#Q5. Episode wise minimum and maximum of sharks invested? 
df.groupby('episode_number')[['total_sharks_invested','deal_amount']].agg(['min','max'])
```

```python
df.head()
```

```python
#Q6. Minimum and Maximum sharks invested in each brand? 
df.groupby('brand_name')[['total_sharks_invested']].agg(['min','max'])
```

```python
#Q7. Brand names who ask for 1 crore and got a deal? (deal amount should crore)
df[(df['pitcher_ask_amount']==100) & (df['deal_amount']>=100)][['brand_name','deal_amount']]
```

```python
#
df[(df['pitcher_ask_amount']==100) & (df['deal']==1)][['brand_name','deal_amount']]
```

```python
#Q8. Brand names where deal equity is between 0 to 50? 
df[(df['deal_equity']>=0) & (df['deal_equity']<=50)][['brand_name','deal_equity']]
```

```python
#Q9. Find the number of brands participated in each episode?
df['brand_name'].nunique()
```

```python
#Q10. How many sharks participated in this show and What were their names?
sharks=[]
for i in df.columns:
    if "present" in i:
        shark=i.split("_")[0]
        sharks.append(shark)
print("No of sharks",len(sharks))
print("list of name of sharks",sharks)
```

```python
df.columns
```

```python
#Q11. Find appearance of each sharks?
a=df[['ashneer_present', 'anupam_present',
       'aman_present', 'namita_present', 'vineeta_present', 'peyush_present',
       'ghazal_present']].sum()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(a.index,a.values)
plt.xticks(rotation=45)
plt.show()
```

### Findings
* Anupam was present in most pitches
* ghazal was present the least number of pitches
